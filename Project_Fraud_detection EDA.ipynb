# Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
leb_encode=LabelEncoder()
```

```python
#Label encoding is done to convert string values to integer.
```

```python
insurance_data['Age']=leb_encode.fit_transform(insurance_data['Age'])
insurance_data['Gender']=leb_encode.fit_transform(insurance_data['Gender'])
insurance_data['Admission_type']=leb_encode.fit_transform(insurance_data['Admission_type'])
insurance_data['apr_drg_description']=leb_encode.fit_transform(insurance_data['apr_drg_description'])
insurance_data['Surg_Description']=leb_encode.fit_transform(insurance_data['Surg_Description'])
insurance_data['Emergency_dept']=leb_encode.fit_transform(insurance_data['Emergency_dept'])
insurance_data['Home_or_self_care']=leb_encode.fit_transform(insurance_data['Home_or_self_care'])
insurance_data['Area_Service']=leb_encode.fit_transform(insurance_data['Area_Service'])
insurance_data['Hospital_County']=leb_encode.fit_transform(insurance_data['Hospital_County'])
insurance_data['Cultural_group']=leb_encode.fit_transform(insurance_data['Cultural_group'])
insurance_data['ethnicity']=leb_encode.fit_transform(insurance_data['ethnicity'])
insurance_data['ccs_diagnosis_code']=leb_encode.fit_transform(insurance_data['ccs_diagnosis_code'])
insurance_data['ccs_procedure_code']=leb_encode.fit_transform(insurance_data['ccs_procedure_code'])
insurance_data['Weight_baby']=leb_encode.fit_transform(insurance_data['Weight_baby'])
insurance_data['Abortion']=leb_encode.fit_transform(insurance_data['Abortion'])
```

```python
insurance_data.to_csv('Insurance_cleaned.csv',index=False)
```
